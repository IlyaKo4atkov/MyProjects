## Задание 1. Подсчет количества букв

**Описание задания.** Написал программу, которая считывает список слов из файла и собирает статистику о количестве встречаемых букв в тексте. Вывел результат для всех 26 букв английского алфавита и отдельно отметил букву, которая встречалась в словах наиболее редко.


```python

with open('story.txt', 'w', encoding='utf-8') as file: # открываем файл для записи (если файл отсутствует в файловой системе, то он создается)
    text = """The Zen of Python, by Tim Peters
Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
Flat is better than nested.
Sparse is better than dense.
Readability counts.
Special cases aren't special enough to break the rules.
Although practicality beats purity.
Errors should never pass silently.
Unless explicitly silenced.
In the face of ambiguity, refuse the temptation to guess.
There should be one-- and preferably only one --obvious way to do it.
Although that way may not be obvious at first unless you're Dutch.
Now is better than never.
Although never is often better than *right* now.
If the implementation is hard to explain, it's a bad idea.
If the implementation is easy to explain, it may be a good idea.
Namespaces are one honking great idea -- let's do more of those!"""
    file.write(text) # записываем в файл текст, хранящийся в переменной text

with open('story.txt', 'r', encoding='utf-8') as read: # открывает файл для чтения
    read_files = list(map(lambda x: x.strip().upper(), read.readlines())) # создаем список с помощью функции map, используя анонимную функцию
    result=[]
    for el in read_files: # циклом перебираем элементы списка
        result.append(el.strip().upper()) # удаляем лишние пробелы и знаки переноса строки и переводим строку в верхний регистр
    print(result)
    counts = {}
    for ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
        counts[ch] = 0 # Для каждой буквы инициализируем счетчик нулем
    for el in result: # проводим итерацию по списку
        for i in el:
            if i not in "'.,!*- ": # проверка символа на отсутствие "'.,!*- "
                counts[i] = counts.get(i, 0) + 1 # в случае нахождения буквы, счетчик прибавляет к значению + 1
    print(counts)
    out = []
    for key, value in counts.items(): # проводим итерацию по словарю как ключ, значение
        print(f'Буква:{key} встречается {value} раз')
        smallest_count = min(counts.values())
        if value == smallest_count:
            out.append(key)
    print(f'Наиболее редко используемые буквы: {", ".join(out)}')
```
